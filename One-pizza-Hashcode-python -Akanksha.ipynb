importing pandas and numpy libraries


```python
import pandas as pd
import numpy as np
```

creating dataframe and reading csv file


```python
df=pd.read_csv("e_elaborate.in.txt")
df
```

seperated even number lines into new dataframe(df2)---(even number indexed lines contains liked ingredients)


```python
df2 = df.iloc[df.index%2==0]
df2.columns = ["like"]
df2.index = np.arange(len(df2))
df2["no_of_like"] = df2["like"].apply(lambda x : int(x.split(" ")[0]))
df2.head()
```

seperated odd number lines into new dataframe(df3)-- (odd indexes contains disliked ingredients)

```python
df3 = df.copy()
df3 = df3.iloc[df.index%2==1]
df3.columns = ["dislike"]
df3.index = np.arange(len(df3))
df3["no_of_dislike"] = df3["dislike"].apply(lambda x : int(x.split(" ")[0]) )
df3.head()
```

merged df2 and df3


```python
df4 = pd.concat([df2,df3],axis = 1)
df4.head()
```

droped those rows which has dislike number greater or equals to 3

```python
indexNames = df4[df4['no_of_dislike'] >= 3].index
df4 = df4.drop(indexNames)
df4.head()
```

created new datafarame df5 with only liked ingredients

```python
df5 = df4.drop(["dislike","no_of_dislike","no_of_like"],axis = 1)
df5.head()
```

created new dataframe df6 with only disliked ingredients

```python
df6 = df4.drop(["like","no_of_like","no_of_dislike"],axis=1)
df6.head()
```

seperated all liked ingredients into separate columns

```python
for n in range(1,6):
  df5["like"+str(n)] = df5["like"].apply(lambda x : x.split(" ")[n] if len(x.split())>n else 0)
```

```python
df5 = df5.drop(["like"],axis=1)
df5.head()
```

unstack the dataframe df5


```python
df5 = df5.unstack()
df5.head()
```

counted how many times particular ingredient comes in df5


```python
df5 = df5.value_counts()
df5.head()
```

```python
df5 = df5.to_frame()
df5.head()
```

```python
df5 = df5.iloc[1: , :]
df5.head()
```

final dataset representing like score of all ingredients

```python
df5 = df5.set_axis(['like_score'], axis=1, inplace=False)
df5.head()
```

separated disliked ingredient into separate columns

```python
for n in range(1,3):
  df6["dislike"+str(n)] = df6["dislike"].apply(lambda x : x.split(" ")[n] if len(x.split())>n else 0)
df6 = df6.drop(["dislike"],axis = 1)
df6.head()
```

```python
df6 = df6.unstack()
df6.head()
```

```python
df6 = df6.value_counts()
df6.head()
```

```python
df6 = df6.to_frame()
df6 = df6.set_axis(['dislike_score'], axis=1, inplace=False)
df6 = df6.iloc[1: , :]
df6.head()
```

final dataset representing total dislike score for all ingredients

merged dataframe of like score and dislike score


```python
df7 = pd.concat([df5,df6],axis=1)
df7.head()
```

replaced null values to 0



```python
df7 = df7.fillna(0)
df7.head()
```

added extra column representing names of ingredients

```python
df7["Ingredient"] = df7.index
df7.head()
```

created list with ingredients which has like score greater then dislike score

```python
Ingre_list = []
for n in range(1,len(df7)):
  if df7.iloc[n,0] >= df7.iloc[n,1]:
    Ingre_list.append(df7.iloc[n,2])
```

```python
len(Ingre_list)
```

hash code submission file created

```python
str1 = ""
for i in Ingre_list:
  str1 =  str1+" "+i 

ans=  str(len(Ingre_list))+" "+str1    
newfile = open("Ans.txt","w+")
newfile.write(ans)
```

# Score Simulator


```python
import linecache

input_filename = "e_elaborate.in.txt"
output_filename="Ans.txt" 


pl = int(linecache.getline(input_filename, 1))
with open(input_filename,'r') as f:  
  list_indi = [[0 for a in range(2)] for b in range(pl)] 
  def find_ingerdiants(list_indi):
    a=0
    for i in f:      
      a+=1
      if len(i)>2:
        k = i.split()
        if (a%2==0):
          k.pop(0)          
          list_indi[(a//2)-1][0]=k          
        else:
          k.pop(0)  
          list_indi[(a//2)-1][1]=k          
           
  find_ingerdiants(list_indi)

input_file = open(output_filename,"r")
lst = input_file.read().split()
lst.pop(0)

customer=0
for i in range(0,pl):
  a=0
  
  if list_indi[i][1] != 0:

    for j in range(0,len(list_indi[i][1])):
      if list_indi[i][1][j] in lst:
        a+=1
    if a>0:
      continue  
  if (set(list_indi[i][0]).issubset(set(lst))):
    customer+=1
    
print(customer)
```
